```python
# let's create a function to find the player's OBP
def calculate_player_obp(H, HBP, BB, AB, SF):
   dividend = H + BB + HBP 
   divisor = AB + BB + HBP + SF
   obp = dividend / divisor 
   return obp
```

```python
# a function that shows per dollar of investment; the fraction of a percent of OBP per dollar spent
def money_ball(OBP, salary):
   return OBP / salary
```
